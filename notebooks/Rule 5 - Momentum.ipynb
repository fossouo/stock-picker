This one is different.  We have to get historical data to measure percent change in stock price plus any dividends that are paid out.

```python
all_stock_prices = get_pricing(security_list, fields="price", start_date=six_months_ago, end_date=today)
```

```python
length_price_data = len(all_stock_prices)
for symbol in all_stock_prices.columns.values:
    # We should gather a couple of things here, some noise or data health indicators
    # -- % NaN
    # -- how many days did we lose up front finding a valid price
    # -- same on the backend
    # -- the actual change in price
    # -- Need to figure out if dividends are actually being reflected here or not
    
    start_price_cursor = 0
    start_price = float('nan')
    while math.isnan(start_price) and start_price_cursor < length_price_data:
        start_price = all_stock_prices[symbol][start_price_cursor]
        start_price_cursor += 1
        
    end_price_cursor = length_price_data - 1
    end_price = float('nan')
    while math.isnan(end_price) and end_price_cursor > start_price_cursor:
        end_price = all_stock_prices[symbol][end_price_cursor]
        end_price_cursor -= 1
        
        
    print symbol, start_price, end_price, ((end_price - start_price) / start_price)
```
